```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel(r'F:\study\Conestoga\Big Data Solution Architecture\Programming for Big Data\Assignments\ASSIGNMENT-4_&_5-OK\ASSIGNMENT-4_&_5-OK\CC-3Y-Transactions.xlsx')
```

```python
data['year']=data['Date'].dt.year
```

```python
data_2013=(data.loc[data['year'].eq(2013)]\
        .groupby('City')['Amount'].agg('sum')\
        .reset_index(name='Total_Expense'))
```

```python
topcity_2013=(data_2013.nlargest(5,['Total_Expense']))
topcity_2013=topcity_2013.sort_values(['Total_Expense'],ascending=False)
y=list(map(str,topcity_2013['Total_Expense'].tolist()))
pd.DataFrame([y],index=['2013'],columns=topcity_2013['City'])
```

```python
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])
axis.pie(topcity_2013['Total_Expense'], labels=topcity_2013['City'],
         autopct='%.1f%%')
axis.set_title('Top 5 City\'s Expense in 2013', size=15, pad=10)
```

```python
data['MM']=pd.DatetimeIndex(data['Date']).month
def func(x):
    if x<=3: 
        return "Quarter1"
    elif x<=6:
        return "Quarter2"
    elif x<=9:
        return "Quarter3"
    else: 
        return "Quarter4"

data['quarter']=data['MM'].apply(func)
```

```python
filter = (data['year'] == 2013)
data[filter].head(5)
data2013 = data[filter].groupby(['year','Gender','quarter'])['Amount'].sum()
data2013
```

```python
legend = ['F:Q4', 'M:Q4']
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])
index = range(len(data2013))

axis.bar(index, data2013)
axis.set_title('Gender based-Expenses (2013)', fontsize=15, pad=10)
axis.set_xlabel('Gender:QuarterNumber', fontsize=15)
axis.set_ylabel('Expense', fontsize=15)
axis.set_xticks(index)
axis.set_xticklabels(legend)
```

```python
#Fraud Detection 
data['Day'] = pd.DatetimeIndex(data['Date']).day
data['DayName'] = pd.DatetimeIndex(data['Date']).day_name()
data['DayofWeek'] = pd.DatetimeIndex(data['Date']).dayofweek
data
```

```python
#Fraud Detection

#1) If day of the transaction is first day of week and expense type is "Entertainment" and expense amount is more than '100000' and
#   Day of month falls between 1-10 of month. It is quite difficult to spend this much amount on first day of week and in first 10 days of
#   Months as salary will might not be credit till that and other expense could be there so its difficult to spend huge amount on this time
#   having silver card. as silver card credit is not much compare to other types of card.

filter = ((data['Exp Type'] == 'Entertainment') & (data['Day'] <= 10) & (data['DayofWeek'] == 0) & (data['Amount'] > 100000) & (data['Card Type'] == 'Silver'))     
df1 = data[filter]
df1
```

```python
#Fraud Detection
#2) If day of the transaction is between "15 to 20" and expense type is "Bills".so it will consider as fraud because people 
#   pay their bills on end of the month or start of month. 

filter2 = ((data['Day'] >= 15) & (data['Day'] <= 20) & (data['Exp Type'] == 'Bills'))           
df2 =data[filter2]
df2
```
